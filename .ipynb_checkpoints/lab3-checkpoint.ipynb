# Подключение базы данных и загрузка датасета

```python
# study = optuna.create_study(study_name = "example-study", storage="postgresql://user:opt@localhost:5432/optdb")
# print(f"Sampler is {study.sampler.__class__.__name__}")

storage="postgresql://user:opt@localhost:5432/optdb"
```

```python
# Загрузка датасета
data = load_diabetes()
X = data.data
y = data.target

# Масштабирование признаков
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Разделение данных на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
```

# Разберем регрессию случайным лесом

### Т.к. в официальной док-ции указано, что лучший pruner для RandomSampler - MedianPruner, то сначала разберем данный вариант, а затем случай с менее подходящим pruner.

### RandomSampler подразумевает случайный выбор гиперпараметров, MedianPruner - сравнивает результат с медианным значением предыдущих экспериментов.

```python
study_rf_reg_rndsmp_median = optuna.create_study(study_name = "random-forest-regressor-randomsampler-median-pruner",
                                                 direction = "maximize",
                                                 storage = storage,
                                                 sampler = optuna.samplers.RandomSampler(),
                                                 pruner = optuna.pruners.MedianPruner())
```

### Второй случай: RandomSampler - HyperbandPruner

### HyperbandPruner сначала раздает малую часть ресурсов всем экспериментам. Эксперименты с плохими результатами прерываются.

```python
study_rf_reg_rndsmp_hpr = optuna.create_study(study_name = "random-forest-regressor-randomsampler-hyperband-pruner",
                                                 direction = "maximize",
                                                 storage = storage,
                                                 sampler = optuna.samplers.RandomSampler(),
                                                 pruner = optuna.pruners.HyperbandPruner())
```

```python
def objective_rf(trial):
    # Гиперпараметры для оптимизации
    n_estimators = trial.suggest_int("n_estimators", 50, 300)
    max_depth = trial.suggest_int("max_depth", 3, 20)
    min_samples_split = trial.suggest_int("min_samples_split", 2, 20)
    min_samples_leaf = trial.suggest_int("min_samples_leaf", 1, 10)

    # Модель RandomForest
    model = RandomForestRegressor(
        n_estimators=n_estimators,
        max_depth=max_depth,
        min_samples_split=min_samples_split,
        min_samples_leaf=min_samples_leaf,
        random_state=42
    )
    # Кросс-валидация
    score = cross_val_score(model, X_train, y_train, cv=3, scoring="neg_mean_squared_error").mean()
    return score 
```

```python
study_rf_reg_rndsmp_median.optimize(objective_rf, n_trials=300)
```

```python
print("Best estimation parameters:", study_rf_reg_rndsmp_median.best_params)
```

```python
study_rf_reg_rndsmp_hpr.optimize(objective_rf, n_trials=300)
```

```python
print("Best estimation parameters:", study_rf_reg_rndsmp_hpr.best_params)
```

```python
optuna.visualization.plot_optimization_history(study_rf_reg_rndsmp_median)
```

```python
optuna.visualization.plot_param_importances(study_rf_reg_rndsmp_median)
```

```python
optuna.visualization.plot_optimization_history(study_rf_reg_rndsmp_hpr)
```

```python
optuna.visualization.plot_param_importances(study_rf_reg_rndsmp_hpr)
```
